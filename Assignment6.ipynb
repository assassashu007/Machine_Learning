```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = os.getcwd()+"\dataset\weather.csv"
data = pd.read_csv(path)
data.head(5)
```

```python
plt.plot(data["MinTemp"])
plt.xlabel("Minimum Temperature")
plt.show()
```

```python
plt.scatter(data["MinTemp"], data["MaxTemp"])
plt.xlabel("Minimum Temprature")
plt.ylabel("Maximum Temperature")
plt.axis([-20,30,0,40])
plt.show()
```

```python
plt.hist(data["MaxTemp"])
plt.show()
```

```python
sns.distplot(data["MaxTemp"])
```

```python
sns.distplot(data["MaxTemp"], hist=False, rug=True)
```

```python
sns.kdeplot(data["Evaporation"], shade=True, color='r', bw=7)
```

```python
sns.jointplot(data["MinTemp"], data["MaxTemp"])
```

```python
sns.jointplot(data["MinTemp"], data["MaxTemp"], kind='hex')
```

```python
sns.jointplot(data["MinTemp"], data["MaxTemp"], kind='kde')
```

```python
sns.lmplot(x="MinTemp",y="MaxTemp",data=data, hue="RainToday")
```

```python
sns.lmplot(x="MinTemp",y="MaxTemp",data=data, hue="RainToday", fit_reg=False)
```

```python
sns.stripplot(x=data["Rainfall"],y=data["Evaporation"], data=data)
```

```python
sns.swarmplot(x=data["Rainfall"],y=data["Evaporation"], data=data)
```

```python
sns.boxplot(x="MinTemp",y="RainToday",data=data)
```

```python
sns.barplot(y="MinTemp",data=data)
```

```python
corr= data.corr()
sns.heatmap(corr)
```
